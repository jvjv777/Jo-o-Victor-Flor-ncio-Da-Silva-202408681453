```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Bibliotecas importadas com sucesso!")
```

```python
# Criando dados simulados de uma livraria
np.random.seed(42)

livros = ["Python Básico", "Data Science", "Machine Learning", "Flask Web", "SQL Essencial"]
categorias = ["Programação", "Dados", "IA", "Web", "Banco de Dados"]

dados = {
    "Livro": np.random.choice(livros, 100),
    "Categoria": np.random.choice(categorias, 100),
    "Preco": np.random.uniform(30, 150, 100).round(2),
    "Quantidade": np.random.randint(1, 10, 100),
}

df = pd.DataFrame(dados)

# Criar coluna faturamento
df["Faturamento"] = df["Preco"] * df["Quantidade"]

# Salvar CSV
df.to_csv("livraria.csv", index=False)

df.head()
```

```python
print("Primeiras linhas:")
print(df.head())

print("\nInformações do dataset:")
print(df.info())

print("\nEstatísticas:")
print(df.describe())
```

```python
# Faturamento total
faturamento_total = df["Faturamento"].sum()
print("Faturamento total:", faturamento_total)

# Faturamento por categoria
faturamento_categoria = df.groupby("Categoria")["Faturamento"].sum()
print("\nFaturamento por categoria:")
print(faturamento_categoria)

# Livro mais vendido
mais_vendido = df.groupby("Livro")["Quantidade"].sum().sort_values(ascending=False)
print("\nLivros mais vendidos:")
print(mais_vendido)
```

```python
# Gráfico de faturamento por categoria
faturamento_categoria.plot(kind="bar")
plt.title("Faturamento por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Faturamento")
plt.xticks(rotation=45)
plt.show()

# Gráfico dos livros mais vendidos
mais_vendido.plot(kind="bar")
plt.title("Livros Mais Vendidos")
plt.xlabel("Livro")
plt.ylabel("Quantidade")
plt.xticks(rotation=45)
plt.show()
```
