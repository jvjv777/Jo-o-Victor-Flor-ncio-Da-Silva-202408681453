```python
import pandas as pd
import numpy as np

# ==================== ATIVIDADE 03 - PANDAS ====================
print("=== ATIVIDADE 03 - ANÁLISE DE LIVROS ===")

# 1. Carregar o dataset (se você tiver o CSV real, use ele. Senão usamos um de exemplo)
# df = pd.read_csv("livros.csv", sep=";")   # ← descomente se tiver o arquivo

# Dataset de exemplo (para entregar hoje mesmo)
dados = {
    "titulo": ["Dom Casmurro", "O Pequeno Príncipe", "Sapiens", "Python para Dados", "Clean Code", "Harry Potter", "Atomic Habits"],
    "autor": ["Machado de Assis", "Antoine de Saint-Exupéry", "Yuval Noah Harari", "João Silva", "Robert Martin", "J.K. Rowling", "James Clear"],
    "isbn": [123456, 234567, 345678, 456789, 567890, 678901, 789012],
    "paginas": [256, 96, 464, 320, 464, 320, 320],
    "ano": [1899, 1943, 2011, 2023, 2008, 1997, 2018]
}
df = pd.DataFrame(dados)
```

```python
# Nível 1
print("\n1. Primeiras 5 linhas:")
print(df.head())

print("\n2. Informações do dataset:")
print(df.info())
print(df.describe())

print("\n3. Valores nulos por coluna:")
print(df.isnull().sum())
```

```python
# Nível 2
df_limpo = df[df["paginas"] > 0].copy()
df_limpo["faixa_paginas"] = df_limpo["paginas"].apply(
    lambda x: "Curto" if x < 150 else "Médio" if x <= 350 else "Longo"
)
df_limpo["decada"] = (df_limpo["ano"] // 10) * 10

print("\n Nível 2 concluído - df_limpo criado")
```

```python
# Nível 3 (bônus)
print("\nMédia de páginas por década:")
print(df_limpo.groupby("decada")["paginas"].mean().round(1))

print("\nTop 5 autores com mais livros:")
print(df_limpo["autor"].value_counts().head(5))
```

```python
# Exportar
df_limpo.to_excel("livros_analisados.xlsx", index=False)
print("\n Arquivo 'livros_analisados.xlsx' gerado!")
```
